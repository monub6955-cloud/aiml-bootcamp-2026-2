```python
import pandas as pd
df = pd.read_csv("session4_dummy_students_data.csv.xls")
df['average'] = df[['math', 'science', 'english']].mean(axis=1)
result_list=[]
for avg in df['average']:
  if avg >=40:
    result_list.append('Pass')
  else:
    result_list.append('Fail')
df['result'] = result_list
class_avg = df['average'].mean()
highest_avg = df['average'].max()
lowest_avg = df['average'].min()
top_students = df[df['average'] == highest_avg]['name'].tolist()
low_attendance_students = df[df['attendance_percent'] < 75]['name'].tolist()
few_assignments_students = df[df['assignments_submitted'] < 6]['name'].tolist()

print(f"Class Average: {class_avg}")
print(f"Highest Average: {highest_avg}")
print(f"Lowest Average: {lowest_avg}")
print(f"Top Student(s): {', '.join(top_students)}")
print(f"Students with attendance below 75%: {', '.join(low_attendance_students)}")
print(f"Students who submitted fewer than 6 assignments: {', '.join(few_assignments_students)}")


df.to_csv("session4_studentreport.csv", index=False)

```

```python
avg_project = df['final_project_score'].mean()
print("Average Project Score:", round(avg_project, 2))

hard_workers = df[df['study_hours_per_week'] > 8]
print("\nStudents studying > 8 hours:")
print(hard_workers[['name', 'study_hours_per_week']])

mumbai_avg = df[df['city'] == 'Mumbai']['average'].mean()
pune_avg = df[df['city'] == 'Pune']['average'].mean()
print(f"\nMumbai Average Marks: {mumbai_avg}")
print(f"Pune Average Marks: {pune_avg}")

df.loc[df['average'] >= 85, 'performance_level'] = 'Excellent'
df.loc[(df['average'] >= 70) & (df['average'] < 85), 'performance_level'] = 'Good'
df.loc[(df['average'] >= 40) & (df['average'] < 70), 'performance_level'] = 'Average'
df.loc[df['average'] < 40, 'performance_level'] = 'Needs Improvement'

new_columns = ['student_id', 'name', 'city', 'average', 'performance_level']
df[new_columns].to_csv("simplified_student_report.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)
print(df.head())
print(df.shape)
print(df.columns)
print(df.info())
```

```python
df=df.rename(columns={'lifeExp':'life_expectancy','pop':'population','gdpPercap':'gdp_per_capita'})

print(df.head())
print(df.columns)
```

```python
print("Missing values in each column:")
print(df.isna().sum)

print("Number of duplicate rows:", df.duplicated().sum())
```

```python
df_clean = df.copy()

# Remove exact duplicate rows if any
if df_clean.duplicated().sum() > 0:
    df_clean = df_clean.drop_duplicates()

# Strip extra spaces from text columns
for col in ["country", "continent"]:
    df_clean[col] = df_clean[col].astype(str).str.strip()

# Ensure expected data types
df_clean["year"] = df_clean["year"].astype(int)
df_clean["population"] = df_clean["population"].astype(float)
df_clean["life_expectancy"] = df_clean["life_expectancy"].astype(float)
df_clean["gdp_per_capita"] = df_clean["gdp_per_capita"].astype(float)

print(df_clean.dtypes)
print(df_clean.head())
```

```python
dirty_demo = df_clean.head(12).copy()

# Deliberately introduce missing values for teaching practice
dirty_demo.loc[2, "life_expectancy"] = np.nan
dirty_demo.loc[5, "gdp_per_capita"] = np.nan

print(dirty_demo)
print(dirty_demo.isna().sum())

# Fill missing numerical values using median
clean_demo = dirty_demo.copy()
clean_demo["life_expectancy"] = clean_demo["life_expectancy"].fillna(clean_demo["life_expectancy"].median())
clean_demo["gdp_per_capita"] = clean_demo["gdp_per_capita"].fillna(clean_demo["gdp_per_capita"].median())

print(clean_demo.isna().sum())
print(clean_demo)
```

```python
# 1. Year range
year_min = df_clean['year'].min()
year_max = df_clean['year'].max()
print(f"Year Range: {year_min} to {year_max}")

# 2. Number of unique countries
num_countries = df_clean['country'].nunique()
print(f"Number of Countries: {num_countries}")

# 3. Number of unique continents
num_continents = df_clean['continent'].nunique()
print(f"Number of Continents: {num_continents}")
```

```python
print(df_clean["continent"].value_counts())

continent_summary=df_clean.groupby("continent")[["life_expectancy","population","gdp_per_capita"]].mean()
print(continent_summary)
#
```

```python
latest_year=df_clean["year"].max()
latest_df=df_clean[df_clean["year"]== latest_year].copy()

print("latest_year:",latest_year)
print(latest_df.head())
print(latest_df.shape)
```

```python
# 1. Group by continent and calculate the average life expectancy
avg_life_exp = df_clean.groupby("continent")["life_expectancy"].mean().reset_index()

# 2. Sort the values from lowest to highest for a cleaner chart layout
avg_life_exp_sorted = avg_life_exp.sort_values(by="life_expectancy", ascending=True)

# 3. Create the bar chart
plt.bar(avg_life_exp_sorted['continent'], avg_life_exp_sorted['life_expectancy'], color='skyblue', edgecolor='black')

# 4. Add titles, labels, and gridlines
plt.title(f'Average Life Expectancy by Continent in {latest_year}')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy ')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.xticks(rotation=45)
plt.tight_layout()
# 5. Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Filter data for India, China, and the US
in_df = df_clean[df_clean["country"] == "India"]
ch_df = df_clean[df_clean["country"] == "China"]
us_df = df_clean[df_clean["country"] == "United States"]

# 2. Plot lines
plt.plot(in_df["year"], in_df["life_expectancy"], marker="o", label="India")
plt.plot(ch_df["year"], ch_df["life_expectancy"], marker="o", label="China")
plt.plot(us_df["year"], us_df["life_expectancy"], marker="o", label="USA")

# 3. Add details and display
plt.title("Life Expectancy Over Time")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Create the scatter plot (GDP on X-axis, Life Expectancy on Y-axis)
plt.scatter(df_clean['gdp_per_capita'], df_clean['life_expectancy'], color='purple', alpha=0.5, edgecolor='white')

# 2. Add labels and title
plt.title('GDP per Capita vs. Life Expectancy')
plt.xlabel('GDP per Capita ($)')
plt.ylabel('Life Expectancy (Years)')
plt.grid(True, linestyle='--', alpha=0.5)

# 3. Display the chart
plt.tight_layout()
plt.show()
```
